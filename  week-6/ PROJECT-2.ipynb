```python
 import tkinter as tk
from tkinter import messagebox

# Lists to store admitted and not-admitted candidates
admitted = []
not_admitted = []

# Function to check admission criteria for Computer Science
def check_computer_science():
    name = name_entry.get()
    jamb_score = int(jamb_entry.get())
    math_grade = math_var.get()
    english_grade = english_var.get()
    chemistry_grade = chemistry_var.get()
    physics_grade = physics_var.get()
    biology_grade = biology_var.get()
    passed_interview = interview_var.get()

    if (jamb_score >= 250 and
        any([grade in ['A1', 'B2', 'B3', 'C4', 'C5', 'C6'] for grade in [math_grade, english_grade, chemistry_grade, physics_grade, biology_grade]]) and
        passed_interview):
        admitted.append(f"{name} - Computer Science")
        messagebox.showinfo("Admission Status", f"Congratulations, {name}! You are admitted into  the department of Computer Science.")
    else:
        not_admitted.append(f"{name} - Computer Science")
        messagebox.showinfo("Admission Status", f"Sorry, {name}. You are not admitted into the department of computer science Computer Science.")

 

# Create main window
root = tk.Tk()
root.title("Admission Checker")
root.geometry("300x500")

# Labels and Entries
name_label = tk.Label(root, text="Full Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

jamb_label = tk.Label(root, text="JAMB Score:")
jamb_label.pack()
jamb_entry = tk.Entry(root)
jamb_entry.pack()

math_label = tk.Label(root, text="Mathematics Grade:")
math_label.pack()
math_var = tk.StringVar(root)
math_var.set("A1")
math_option = tk.OptionMenu(root, math_var, "A1", "B2", "B3", "C4", "C5", "C6","D7","E8","F9")
math_option.pack()

english_label = tk.Label(root, text="English Grade:")
english_label.pack()
english_var = tk.StringVar(root)
english_var.set("A1")
english_option = tk.OptionMenu(root, english_var, "A1", "B2", "B3", "C4", "C5", "C6","D7","E8","F9")
english_option.pack()

chemistry_label = tk.Label(root, text="Chemistry Grade:")
chemistry_label.pack()
chemistry_var = tk.StringVar(root)
chemistry_var.set("A1")
chemistry_option = tk.OptionMenu(root, chemistry_var, "A1", "B2", "B3", "C4", "C5", "C6","D7","E8","F9")
chemistry_option.pack()

physics_label = tk.Label(root, text="Physics Grade:")
physics_label.pack()
physics_var = tk.StringVar(root)
physics_var.set("A1")
physics_option = tk.OptionMenu(root, physics_var, "A1", "B2", "B3", "C4", "C5", "C6","D7","E8","F9")
physics_option.pack()

biology_label = tk.Label(root, text="Biology Grade:")
biology_label.pack()
biology_var = tk.StringVar(root)
biology_var.set("A1")
biology_option = tk.OptionMenu(root, biology_var, "A1", "B2", "B3", "C4", "C5", "C6","D7","E8","F9")
biology_option.pack()


# Label for the interview question
interview_label = tk.Label(root, text="Did you pass the interview?:")
interview_label.pack()

# BooleanVar to store interview status
interview_var = tk.BooleanVar()

# Display button for "Yes" option
yes_button = tk.Radiobutton(root, text="Yes", variable=interview_var, value=True)
yes_button.pack()

# display button for "No" option
no_button = tk.Radiobutton(root, text="No", variable=interview_var, value=False)
no_button.pack()

department_label = tk.Label(root, text="Department:")
department_label.pack()
department_var = tk.StringVar(root)
department_var.set("Computer Science")
department_option = tk.OptionMenu(root, department_var, "Computer Science")
department_option.pack()

# Buttons to check admission
computer_science_button = tk.Button(root, text="Check Computer Science Admission", command=check_computer_science)
computer_science_button.pack()

root.mainloop()

# Print the lists of admitted and not-admitted candidates
print("Admitted Candidates:")
for candidate in admitted:
    print(candidate)

print("\nNot Admitted Candidates:")
for candidate in not_admitted:
    print(candidate)


 


```
